```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Binary Step Activation Function
## It returns '0' if the input is the less then zero
## It returns '1' if the input is greater than zero
```

```python
def binaryStep(x):
    return np.heaviside(x,1)
```

```python
x = np.linspace(-10, 10)
plt.plot(x, binaryStep(x))
plt.axis('tight')
plt.title('Activation Function :binaryStep')
plt.show()
```

```python
# Linear Activation Function
# y = f(x) It returns the input as it is

```

```python
def linear(x):
    return x
```

```python
x = np.linspace(-10, 10)
plt.plot(x, linear(x))
plt.axis('tight')
plt.title('Activation Function :Linear')
plt.show()
```

```python
# Sigmoid Activation Function
# It returns 1/(1+exp(-x)). where the values lies between zero and one 
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
# Tanh Activation Function
# Tanh outputs between -1 and 1. Tanh also suffers from gradient problem near the boundaries 
#just as Sigmoid activation function does.
#It returns the value (1-exp(-2x))/(1+exp(-2x)) and the value returned will be lies in between -1 to 1.
```

```python
def tanh(x):
    return np.tanh(x)
```

```python
x = np.linspace(-10, 10)
plt.plot(x, tanh(x))
plt.axis('tight')
plt.title('Activation Function :Tanh')
plt.show()
```

```python
# RELU Activation Function
# RELU is more well known activation function which is used in the deep learning networks. 
# RELU is less computational expensive than the other non linear activation functions.

# RELU returns 0 if the x (input) is less than 0
# RELU returns x if the x (input) is greater than 0
```

```python
def RELU(x):
    x1=[]
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)

    return x1
```

```python
x = np.linspace(-10, 10)
plt.plot(x, RELU(x))
plt.axis('tight')
plt.title('Activation Function :RELU')
plt.show()
```
